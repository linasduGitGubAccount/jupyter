# Tensorflow之神经网络

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# 获取真实的数据
mnist=input_data.read_data_sets('Data\day_6',one_hot=True)
```

```python
# 1、建立占位符 x[None,784] y_true[None,784]
with tf.variable_scope('data'):
    x=tf.placeholder(tf.float32,[None,784])

    y_true=tf.placeholder(tf.int32,[None,10])
```

```python
# 2、建立一个全连接层的神经网络 w[784,10] b[10]
with tf.variable_scope('fc_model'):
    # 随机初始化权重和偏置
    weight=tf.Variable(tf.random_normal([784,10],mean=0.0,stddev=1.0),name='w')

    bias=tf.Variable(tf.constant(0.0,shape=[10]))

    # 预测None个样本的输出结果
    y_predict=tf.matmul(x,weight)+bias
```

```python
# 3、求出所有样本损失，然后求平均值
with tf.variable_scope('soft_cross'):
    # 求平均交叉熵损失
    loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_true,logits=y_predict))
```

```python
# 4、由梯度下降求出损失
with tf.variable_scope('optimizer'):
    train_op=tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

```python
# 5、计算准确率
with tf.variable_scope('acc'):
    equal_list=tf.equal(tf.argmax(y_true,1),tf.argmax(y_predict,1))

    accuracy=tf.reduce_mean(tf.cast(equal_list,tf.float32))
```

```python
# 收集变量,单个数字的收集
tf.summary.scalar('losses',loss)
tf.summary.scalar('acc',accuracy)

# 高纬度变量收集
tf.summary.histogram('weightes',weight)
tf.summary.histogram('biases',bias)
```

```python
# 定义一个合并变量
merged=tf.summary.merge_all()
```

```python
# 创建一个Saver
saver=tf.train.Saver()
```

```python
# 定义一个初始化变量的op
init_op=tf.global_variables_initializer()

# 开启会话训练
with tf.Session() as sess:

    # 初始化变量
    sess.run(init_op)

    # 建立events文件，然后写入
    filewriter=tf.summary.FileWriter('./Data/tem/',graph=sess.graph)

    # 迭代步数去训练 进行预测
    for i in range(2000):

        mnist_x,mnist_y=mnist.train.next_batch(50)

        # 运行train_op
        sess.run(train_op,feed_dict={x:mnist_x,y_true:mnist_y})

        # 写入每部训练的值
        summary=sess.run(merged,feed_dict={x:mnist_x,y_true:mnist_y})

        filewriter.add_summary(summary,i)

        print('训练%d次，准确率为%f：' % (i,sess.run(accuracy,feed_dict={x:mnist_x,y_true:mnist_y})))
    
    # 保存模型
    saver.save(sess,'./Data/tem/ckpt/fc_model')
```
