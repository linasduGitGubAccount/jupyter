```python
import numpy as np 
import pandas as pd
```

```python
movie=pd.read_csv('Data\pandas\IMDB-Movie-Data.csv')
```

```python
movie.head()
```

```python
movie['Genre'].head()
```

```python
temp_list=[i.split(',') for i in movie['Genre']]
```

```python
array_list=np.unique([j for i in temp_list for j in i])
array_list
```

```python
array_list.shape
```

```python
movie.shape
```

```python
# 创建一个全为0的矩阵
genre_zero=pd.DataFrame(np.zeros([1000,20]),columns=array_list)
genre_zero.head()
```

```python
for i in range(1000):
    temp=[i for i in temp_list[i]]
    genre_zero.loc[i,temp]=1
```

```python
genre_zero
```

```python
genre_zero.sum().sort_values(ascending=False).plot(kind='bar')
```
