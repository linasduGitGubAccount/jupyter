# 可视化学习之seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

```python
sns.set_style('whitegrid')
data=np.random.normal(size=(20,6))+np.arange(6)/2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
# x轴y轴加了一些线段
sns.set_style('ticks')
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
sns.violinplot(data)
sns.despine(offset=10) # 指定图与轴线的距离
```

```python
sns.set_style('whitegrid')
sns.boxplot(data=data,palette='deep')
sns.despine(left=True) # 隐藏左边的轴
```

## seanborn的调色板  
* color_palette()能传入任何matplotlib里面的颜色  
* color_palette()不屑参数则默认颜色  
* set_palette()设置所有图的颜色

```python
# 分类色板,默认深色
current_palette=sns.color_palette()
sns.palplot(current_palette)
```

### 圆形画板  
当有六个以上的分类需要分区时，最简单的方法就是在一个圆形空间上均匀的分割颜色  
最常用的方法就是hls的颜色空间

```python
sns.palplot(sns.color_palette('hls',12))
```

```python
# 使用颜色
data=np.random.normal(size=(20,8))+np.arange(8)/2
sns.boxplot(data=data,palette=sns.color_palette('hls',8))
```

hls_palette()函数来控制颜色亮度和饱和度  
* l:亮度，lightness  
* s:饱和度，saturation

```python
sns.palplot(sns.hls_palette(12,l=0.7,s=0.8))
```

```python
# 现在调出的调色板是成对存在的
sns.palplot(sns.color_palette('Paired',12))
```

### 使用xkcd颜色来命名颜色(需要了解命名方式)

```python
plt.plot([0,1],[0,1],sns.xkcd_rgb['pale red'],lw=3)
plt.plot([0,1],[0,2],sns.xkcd_rgb['medium green'],lw=3)
plt.plot([0,1],[0,3],sns.xkcd_rgb['denim blue'],lw=3)
```

### cubehelix_palette()调色板

```python
sns.palplot(sns.color_palette('cubehelix',8))
```

```python
sns.palplot(sns.cubehelix_palette(8,start=0.5,rot=-0.75))
```

### light_palette()和dark_palette()调用定制连续调色板

```python
sns.palplot(sns.light_palette('green'))
```

```python
sns.palplot(sns.dark_palette('purple',n_colors=12))
```

```python
sns.palplot(sns.dark_palette('purple',n_colors=12,reverse=True))
```
